```python
# import packages
import selenium
from selenium import webdriver
```

```python
# path of chromedriver
path = "C:/Users/Manthan/Downloads/chromedriver.exe"
```

```python
# setting the driver
driver = webdriver.Chrome(path)
```

```python
# opens page with driver 
driver.get("https://www.propertiesguru.com/residential-search/2bhk-residential_apartment_flat-for-sale-in-new_delhi")
```

```python
# driver.close()  # to close webpage
# driver.quit()   # to close entire browser
```

```python
# Title of webpage
print(driver.title)
```

```python
# Full source code of page
print(driver.page_source)
```

```python
# All text data and printing it
main = driver.find_element_by_id("searchresultview")
print(main.text)
```

```python
# checking datatype
type(main.text)
```

```python
# extracting the textual data from page
import pandas as pd
s = main.text
print(s)
```

```python
# tokenization of scrapped data string
t = s.split('\n')
print(t)
```

```python
# creating dataframe to store all properties on site
data = [[t[7],t[8],t[10],t[12],t[14],t[15],t[16],t[17],t[18],t[19],t[20]]]
df = pd.DataFrame(data, columns=['Location','Price @rate','Area','Facing','Status','Floor','Furniture','Feature','Bathrooms','Agent','Time'])
df
```

```python
# appending all rows 
for i in range(23):
    n = 16*i+23
    data = [t[n],t[n+1],t[n+3],t[n+5],t[n+7],t[n+8],t[n+9],t[n+10],t[n+11],t[n+12],t[n+13]]
    dflo = len(df)
    df.loc[dflo] = data
```

```python
# making output csv file from dataframe
df.to_csv('2bhk_houses.csv')
```
